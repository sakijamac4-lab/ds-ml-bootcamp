```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.tree import DecisionTreeRegressor

# 1. Load Dataset
CSV_PATH=r"D:\Downloads\DS-ML COURSE\clean_car_dataset.csv"
df=pd.read_csv(CSV_PATH)

# 2. Prepare Features & Target
x_features=df.drop(columns=["Price", "LogPrice"])
y_target=df["Price"]

# 3. Split Data
x_train,x_test,y_train,y_test=train_test_split(x_features,y_target,test_size=0.2, random_state=42)

# 4. Train Models
linear_reg=LinearRegression()
linear_reg.fit(x_train, y_train)
linear_reg_predict=linear_reg.predict(x_test)

rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train, y_train)
rf_predictor=rf.predict(x_test)

d_tree=DecisionTreeRegressor(random_state=42)
d_tree.fit(x_train, y_train)
d_tree_predictor=d_tree.predict(x_test)

# 5. Evaluate Performance
def metrics_fun(name, y_actual, y_predict):
    mae=mean_absolute_error(y_actual, y_predict)
    mse=mean_squared_error(y_actual, y_predict)
    rmse=np.sqrt(mse)
    r2=r2_score(y_actual, y_predict)

    print(f"\n{name}: metrics:")
    print(f"mae : {mae:.0f}")
    print(f"mse : {mse:.0f}")
    print(f"rmse : {rmse:.0f}")
    print(f"r2 : {r2:.2f}")

metrics_fun("Linear regression", y_test,linear_reg_predict)
metrics_fun("Random Forest", y_test,rf_predictor)
metrics_fun("Decision Tree", y_test,d_tree_predictor)

# 6. Sanity Check
i=8
one_df_feature=x_test.iloc[[i]]
y_actual=y_test.iloc[i]

lr_one_df_predictor=float(linear_reg.predict(one_df_feature)[0])
rf_one_df_predictor=float(rf.predict(one_df_feature)[0])
dtree_one_df_predictor=float(d_tree.predict(one_df_feature)[0])
print("\none sample sanity-check:")
print(f"Actual price: {y_actual: .0f}")
print(f"Linear regression: {lr_one_df_predictor: .0f}")
print(f"Random Forest: {rf_one_df_predictor: .0f}")
print(f"Decision Tree: {dtree_one_df_predictor: .0f}")
```
