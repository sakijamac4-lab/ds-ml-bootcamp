## 1. Load Dataset


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

df = pd.read_csv('clean_loan_dataset.csv')
df.head()
```

## 2. Prepare Features & Target

```python
y = df["Approved"]
X = df.drop(columns=["Approved"])

print("Features:", list(X.columns))
print("Target: Approved")
print("X shape:", X.shape, "| y shape:", y.shape)
```

## 3. Split Data 

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape :", X_test.shape)
```

## 4. Train Models (Reproduce Lesson)

```python
# Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

print("Logistic Regression and Random Forest trained.")
```

## 5. Research & Train a Third Classifier

**Chosen algorithm: K-Nearest Neighbors (KNN)**

I chose KNN because it makes no assumptions about the underlying distribution of the data and instead classifies a loan applicant based on how similar they are to applicants who received a known outcome. Since the dataset has already been cleaned and scaled (all numeric columns are standardized), KNN is a natural fit — it is distance-based, so having features on comparable scales is important, and this preprocessing step was already done in Part B1. KNN is also easy to interpret to a loan officer ("this applicant looks like these 5 previously-approved/denied applicants"), which is a nice property for a decision that affects real people.

**Source used for research:** scikit-learn's official documentation on `sklearn.neighbors.KNeighborsClassifier` (https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html), which explains the algorithm's parameters (e.g. `n_neighbors`, distance metric) and default behavior.

```python
# K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

print("KNN trained.")
```

## 6. Evaluate Performance

```python
def evaluate_model(name, model, X_test, y_test):
    """Print Accuracy, Precision, Recall, F1-Score (positive class = Approved = 1)
    and the confusion matrix for a fitted classifier."""
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, pos_label=1)
    rec = recall_score(y_test, y_pred, pos_label=1)
    f1 = f1_score(y_test, y_pred, pos_label=1)

    print(f"{name} Performance:")
    print(f"  Accuracy : {acc:.3f}")
    print(f"  Precision: {prec:.3f}  (positive = Approved=1)")
    print(f"  Recall   : {rec:.3f}  (positive = Approved=1)")
    print(f"  F1-Score : {f1:.3f}  (positive = Approved=1)")
    print()

  

    return y_pred
```
