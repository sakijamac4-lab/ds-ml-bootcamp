```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 1. loading dataset
df = pd.read_csv("clean_car_dataset.csv")
```

```python
#2. preparing data feature and target

x = df.drop("Price", axis=1)
y = df["Price"]
```

```python
# 3. Split Data train test and split
x_train , x_test, y_train , y_test = train_test_split(x,y, test_size = 0.2 ,random_state=42)
```

```python
# 4. Train Models and also predicting
# linear regression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

linear_pred = linear_model.predict(x_test)

```

```python
# RondomForestRegressor

reg_model = RandomForestRegressor( n_estimators=100, random_state=42)
reg_model.fit(x_train, y_train)

rf_pred = reg_model.predict(x_test)
```

```python
# 5.Evaluate Performance

# Linear evaluate
linear_r2 = r2_score(y_test, linear_pred)
linear_mae = mean_absolute_error(y_test, linear_pred)
linear_mse = mean_squared_error(y_test, linear_pred)
linear_rmse = np.sqrt(linear_mse)



# regrosser evaluate
rf_r2 = r2_score(y_test, rf_pred)
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)


# print linear
print("Linear Regression Performance:")
print("R2:", linear_r2)
print("MAE:", linear_mae)
print("MSE:", linear_mse)
print("RMSE:", linear_rmse)


# print regressor

print("\nRandom Forest Regression")
print("R2:", rf_r2)
print("MAE:", rf_mae)
print("MSE:", rf_mse)
print("RMSE:", rf_rmse)
```

```python
# 6.Sanity Check
i= 4

x_test_4 = x_test.iloc[[i]]
y_for_sure = y_test.iloc[i]

lr_predict = linear_model.predict(x_test_4)[0]
rf_predict = reg_model.predict(x_test_4)[0]

print(f"the actual price is: {y_for_sure}")
print(f"linear regration prediction is: {lr_predict}")
print(f"Random forest regrossor prediction is: {rf_predict}")
```
