```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

CSV_PATH = "./raw_loan_dataset.csv"
df = pd.read_csv(CSV_PATH)
# df.head()
```

```python
RANDOM_STATE = 42

# 1) Load the cleaned dataset
# --------------------------------
CSV_PATH =  "./raw_loan_dataset.csv"
df = pd.read_csv(CSV_PATH)
# df.head()
```

```python
# 2) Split features (X) and target (y)
# Features = applicant info; label = Approved (1) or Rejected (0)
X = df.drop(columns=["Approved"])
y = df["Approved"]
# df.head()
```

```python
# 3) Train/Test Split
# --------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y,
)

print("=== SPLIT SIZES ===")
print(f"Train: {X_train.shape[0]}")
print(f"Test : {X_test.shape[0]}")
```

```python
 # 4) Train classifiers
# --------------------------------
log_reg = LogisticRegression(
    max_iter=1000,
    random_state=RANDOM_STATE,
)

rf = RandomForestClassifier(
    n_estimators=1000,
    random_state=RANDOM_STATE,
)

decision_tree = DecisionTreeClassifier(
    random_state=RANDOM_STATE,
)

log_reg.fit(X_train, y_train)
rf.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
```

```python
def print_clf_metrics(name, y_true, y_pred):

    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"\n{name}")
    print("-" * 35)
    print(f"Accuracy : {acc:.3f}")
     print(f"Precision: {prec:.3f}")
    print(f"Recall   : {rec:.3f}")
    print(f"F1 Score : {f1:.3f}")


def print_confmat(name, y_true, y_pred):

    cm = confusion_matrix(y_true, y_pred)

    cm_df = pd.DataFrame(
        cm,
        index=[
            "Actual Rejected",
            "Actual Approved",
               ],
        columns=[
            "Pred Rejected",
            "Pred Approved",
        ],
    )

    print(f"\n{name} Confusion Matrix")
    print(cm_df)


def label_str(v):
    return "Approved (1)" if v == 1 else "Rejected (0)"
```

```python
# 6) Show results for both models

log_reg_preds = log_reg.predict(X_test)

print_clf_metrics(
    "Logistic Regression",
    y_test,
    log_reg_preds,
)

print_confmat(
    "Logistic Regression",
    y_test,
    log_reg_preds,
)
```

```python
# 7) Random Forest
# --------------------------------
rf_preds = rf.predict(X_test)

print_clf_metrics(
    "Random Forest",
    y_test,
    rf_preds,
)

print_confmat(
    "Random Forest",
    y_test,
    rf_preds,
)
```

```python
# 8) Decision Tree
# --------------------------------
dt_preds = decision_tree.predict(X_test)

print_clf_metrics(
    "Decision Tree",
    y_test,
    dt_preds,
)

print_confmat(
    "Decision Tree",
    y_test,
    dt_preds,
)
```

```python
# 9) Single Application Check
# --------------------------------
i = 2

x_one_df = X_test.iloc[[i]]
y_true = y_test.iloc[i]

print("\n=== SINGLE APPLICATION CHECK ===")

print(f"Actual Label         : {label_str(y_true)}")

print(
    f"Logistic Regression  : {label_str(int(log_reg.predict(x_one_df)[0]))}"
)

print(
    f"Random Forest        : {label_str(int(rf.predict(x_one_df)[0]))}"
    )

print(
    f"Decision Tree        : {label_str(int(decision_tree.predict(x_one_df)[0]))}"
)
```
