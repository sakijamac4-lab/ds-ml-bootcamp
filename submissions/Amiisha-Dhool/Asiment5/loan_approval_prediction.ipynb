```python

#  IMPORT LIBRARIES

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)


# 2. LOAD CLEAN DATASET

df = pd.read_csv("clean_loan_dataset.csv")

print("Dataset Preview:")
print(df.head())

print("\nInfo:")
print(df.info())

print("\nTarget distribution:")
print(df["Approved"].value_counts())


# 3. FEATURES & TARGET

X = df.drop("Approved", axis=1)
y = df["Approved"]


# 4. TRAIN TEST SPLIT

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nTrain shape:", X_train.shape)
print("Test shape:", X_test.shape)


# 5. MODELS

lr = LogisticRegression(max_iter=1000, random_state=42)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)


# 6. TRAIN MODELS

lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
knn.fit(X_train, y_train)


# 7. EVALUATION FUNCTION

def evaluate(model, name):
    pred = model.predict(X_test)

    print(f"\n======================")
    print(f"{name} Performance")
    print(f"======================")

    print("Accuracy :", round(accuracy_score(y_test, pred), 3))
    print("Precision:", round(precision_score(y_test, pred), 3))
    print("Recall   :", round(recall_score(y_test, pred), 3))
    print("F1-Score :", round(f1_score(y_test, pred), 3))

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, pred))


# 8. MODEL RESULTS

evaluate(lr, "Logistic Regression")
evaluate(rf, "Random Forest")
evaluate(knn, "KNN (Third Model)")


# 9. SANITY CHECK

i = 5  # change if you want

sample = X_test.iloc[[i]]

print("\n")
print("SANITY CHECK")
print("======================")

print("Actual:", y_test.iloc[i])

print("Logistic Regression:", lr.predict(sample)[0])
print("Random Forest:", rf.predict(sample)[0])
print("KNN:", knn.predict(sample)[0])
```
