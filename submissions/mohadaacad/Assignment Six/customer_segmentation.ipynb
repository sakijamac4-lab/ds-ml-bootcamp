```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
```

```python
df = pd.read_csv("C:/Users/mohad/Downloads/raw_wholesale_customers.csv")

print(df.head())

print()

print(df.info())
```

```python
features = [
    "Fresh",
    "Milk",
    "Grocery",
    "Frozen",
    "Detergents_Paper",
    "Delicassen"
]

X = df[features].copy()

print(X.head())
```

```python
for col in features:

    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    X[col] = X[col].clip(lower, upper)

    print(col)
    print("Lower Bound:", lower)
    print("Upper Bound:", upper)
    print()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print("Scaled Shape:", X_scaled.shape)

print()

print(X_scaled[:5])
```

```python
sse = []

for k in range(1,11):

    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init="auto"
    )

    model.fit(X_scaled)

    sse.append(model.inertia_)

print("SSE Values")

for k,value in zip(range(1,11),sse):

    print(f"k = {k}  -->  SSE = {value:.2f}")
```

```python
kmeans = KMeans(
    n_clusters=5,
    random_state=42,
    n_init="auto"
)

kmeans.fit(X_scaled)

df["Cluster"] = kmeans.labels_

print(df.head())
```

```python
silhouette = silhouette_score(
    X_scaled,
    kmeans.labels_
)

dbi = davies_bouldin_score(
    X_scaled,
    kmeans.labels_
)

print("Silhouette Score:", silhouette)

print("Davies-Bouldin Index:", dbi)
```

```python
agg = AgglomerativeClustering(
    n_clusters=5
)

agg_labels = agg.fit_predict(X_scaled)

df["AgglomerativeCluster"] = agg_labels

print(df.head())
```

```python
kmeans_score = silhouette_score(
    X_scaled,
    kmeans.labels_
)

agg_score = silhouette_score(
    X_scaled,
    agg_labels
)

print("K-Means Silhouette Score")

print(kmeans_score)

print()

print("Agglomerative Silhouette Score")

print(agg_score)
```

```python
if kmeans_score > agg_score:

    print("K-Means produced better separated clusters.")

else:

    print("Agglomerative Clustering produced better separated clusters.")
```
